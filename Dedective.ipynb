```python
from PIL import Image, ImageDraw
import random
import numpy as np
import matplotlib.pyplot as plt


# Load the original image
image_path = "./P180-IMG_4314.png"  # Replace this with your image path
original_image = Image.open(image_path).convert("RGBA")

# Function to add a scratch to the image
def add_scratch(image):
    draw = ImageDraw.Draw(image)
    x1 = random.randint(0, image.width)
    y1 = random.randint(0, image.height)
    x2 = random.randint(0, image.width)
    y2 = random.randint(0, image.height)
    draw.line((x1, y1, x2, y2), fill=(0, 0, 0, 255), width=3)

# Function to add a scuff (small dots) to the image
def add_scuff(image):
    draw = ImageDraw.Draw(image)
    x = random.randint(0, image.width)
    y = random.randint(0, image.height)
    draw.point((x, y), fill=(0, 0, 0, 255))

# Function to simulate paint chipping (remove color in a small area)
def add_paint_chipping(image):
    draw = ImageDraw.Draw(image)
    x1 = random.randint(0, image.width)
    y1 = random.randint(0, image.height)
    x2 = x1 + random.randint(5, 15)
    y2 = y1 + random.randint(5, 15)
    draw.rectangle((x1, y1, x2, y2), fill=(255, 255, 255, 255))

# Create copies of the original image to modify
image_with_scratch = original_image.copy()
image_with_scuff = original_image.copy()
image_with_paint_chipping = original_image.copy()

# Apply the defects
add_scratch(image_with_scratch)
add_scuff(image_with_scuff)
add_paint_chipping(image_with_paint_chipping)

# Show the images
fig, axs = plt.subplots(1, 4, figsize=(20, 5))
axs[0].imshow(np.array(original_image))
axs[0].set_title('Original')
axs[0].axis('off')

axs[1].imshow(np.array(image_with_scratch))
axs[1].set_title('With Scratch')
axs[1].axis('off')

axs[2].imshow(np.array(image_with_scuff))
axs[2].set_title('With Scuff')
axs[2].axis('off')

axs[3].imshow(np.array(image_with_paint_chipping))
axs[3].set_title('With Paint Chipping')
axs[3].axis('off')

plt.show()
original_image = Image.open(image_path).convert("RGBA")

# Function to add a scratch to the image
def add_scratch(image):
    draw = ImageDraw.Draw(image)
    x1 = random.randint(0, image.width)
    y1 = random.randint(0, image.height)
    x2 = random.randint(0, image.width)
    y2 = random.randint(0, image.height)
    draw.line((x1, y1, x2, y2), fill=(0, 0, 0, 255), width=3)

# Function to add a scuff (small dots) to the image
def add_scuff(image):
    draw = ImageDraw.Draw(image)
    x = random.randint(0, image.width)
    y = random.randint(0, image.height)
    draw.point((x, y), fill=(0, 0, 0, 255))

# Function to simulate paint chipping (remove color in a small area)
def add_paint_chipping(image):
    draw = ImageDraw.Draw(image)
    x1 = random.randint(0, image.width)
    y1 = random.randint(0, image.height)
    x2 = x1 + random.randint(5, 15)
    y2 = y1 + random.randint(5, 15)
    draw.rectangle((x1, y1, x2, y2), fill=(255, 255, 255, 255))

# Create copies of the original image to modify
image_with_scratch = original_image.copy()
image_with_scuff = original_image.copy()
image_with_paint_chipping = original_image.copy()

# Apply the defects
add_scratch(image_with_scratch)
add_scuff(image_with_scuff)
add_paint_chipping(image_with_paint_chipping)

# Show the images
fig, axs = plt.subplots(1, 4, figsize=(20, 5))
axs[0].imshow(np.array(original_image))
axs[0].set_title('Original')
axs[0].axis('off')

axs[1].imshow(np.array(image_with_scratch))
axs[1].set_title('With Scratch')
axs[1].axis('off')

axs[2].imshow(np.array(image_with_scuff))
axs[2].set_title('With Scuff')
axs[2].axis('off')

axs[3].imshow(np.array(image_with_paint_chipping))
axs[3].set_title('With Paint Chipping')
axs[3].axis('off')

plt.show()
original_image = Image.open(image_path).convert("RGBA")

# Function to add a scratch to the image
def add_scratch(image):
    draw = ImageDraw.Draw(image)
    x1 = random.randint(0, image.width)
    y1 = random.randint(0, image.height)
    x2 = random.randint(0, image.width)
    y2 = random.randint(0, image.height)
    draw.line((x1, y1, x2, y2), fill=(0, 0, 0, 255), width=3)

# Function to add a scuff (small dots) to the image
def add_scuff(image):
    draw = ImageDraw.Draw(image)
    x = random.randint(0, image.width)
    y = random.randint(0, image.height)
    draw.point((x, y), fill=(0, 0, 0, 255))

# Function to simulate paint chipping (remove color in a small area)
def add_paint_chipping(image):
    draw = ImageDraw.Draw(image)
    x1 = random.randint(0, image.width)
    y1 = random.randint(0, image.height)
    x2 = x1 + random.randint(5, 15)
    y2 = y1 + random.randint(5, 15)
    draw.rectangle((x1, y1, x2, y2), fill=(255, 255, 255, 255))

# Create copies of the original image to modify
image_with_scratch = original_image.copy()
image_with_scuff = original_image.copy()
image_with_paint_chipping = original_image.copy()

# Apply the defects
add_scratch(image_with_scratch)
add_scuff(image_with_scuff)
add_paint_chipping(image_with_paint_chipping)

# Show the images
fig, axs = plt.subplots(1, 4, figsize=(20, 5))
axs[0].imshow(np.array(original_image))
axs[0].set_title('Original')
axs[0].axis('off')

axs[1].imshow(np.array(image_with_scratch))
axs[1].set_title('With Scratch')
axs[1].axis('off')

axs[2].imshow(np.array(image_with_scuff))
axs[2].set_title('With Scuff')
axs[2].axis('off')

axs[3].imshow(np.array(image_with_paint_chipping))
axs[3].set_title('With Paint Chipping')
axs[3].axis('off')

plt.show()
```

# Adjustable Proto2

```python
from PIL import Image, ImageDraw
from ipywidgets import interact, widgets
import random
import numpy as np
import matplotlib.pyplot as plt

# Load the original image for demonstration
image_path = "./P180-IMG_4314.png"  # Replace this with your image path
original_image = Image.open(image_path).convert("RGBA")

# Function to add a scratch to the image
def add_scratch(image, width=3):
    draw = ImageDraw.Draw(image)
    x1 = random.randint(0, image.width)
    y1 = random.randint(0, image.height)
    x2 = random.randint(0, image.width)
    y2 = random.randint(0, image.height)
    draw.line((x1, y1, x2, y2), fill=(0, 0, 0, 255), width=width)

# Function to add a scuff (small dots) to the image
def add_scuff(image, num_dots=1):
    draw = ImageDraw.Draw(image)
    for _ in range(num_dots):
        x = random.randint(0, image.width)
        y = random.randint(0, image.height)
        draw.point((x, y), fill=(0, 0, 0, 255))

# Function to simulate paint chipping (remove color in a small area)
def add_paint_chipping(image, size=(5, 15)):
    draw = ImageDraw.Draw(image)
    x1 = random.randint(0, image.width)
    y1 = random.randint(0, image.height)
    x2 = x1 + random.randint(size[0], size[1])
    y2 = y1 + random.randint(size[0], size[1])
    draw.rectangle((x1, y1, x2, y2), fill=(255, 255, 255, 255))

# Function to apply defects interactively
def apply_defects(scratch_width=3, scuff_dots=1, chip_size=15):
    image_with_scratch = original_image.copy()
    image_with_scuff = original_image.copy()
    image_with_paint_chipping = original_image.copy()

    add_scratch(image_with_scratch, width=scratch_width)
    add_scuff(image_with_scuff, num_dots=scuff_dots)
    add_paint_chipping(image_with_paint_chipping, size=(5, chip_size))

    fig, axs = plt.subplots(1, 4, figsize=(20, 5))
    axs[0].imshow(np.array(original_image))
    axs[0].set_title('Original')
    axs[0].axis('off')

    axs[1].imshow(np.array(image_with_scratch))
    axs[1].set_title('With Scratch')
    axs[1].axis('off')

    axs[2].imshow(np.array(image_with_scuff))
    axs[2].set_title('With Scuff')
    axs[2].axis('off')

    axs[3].imshow(np.array(image_with_paint_chipping))
    axs[3].set_title('With Paint Chipping')
    axs[3].axis('off')

    plt.show()

# Create interactive widgets
interact(apply_defects,
         scratch_width=widgets.IntSlider(min=1, max=10, step=1, value=3),
         scuff_dots=widgets.IntSlider(min=1, max=10, step=1, value=1),
         chip_size=widgets.IntSlider(min=5, max=20, step=1, value=15));
```

# Proto3

```python
from PIL import Image, ImageDraw
from ipywidgets import interact, widgets
from IPython.display import display, clear_output
import random
import numpy as np
import matplotlib.pyplot as plt

# Load the original image
image_path = "./P180-IMG_4314.png"  # Replace this with your image path
original_image = Image.open(image_path).convert("RGBA")

# Initialize output widget
output = widgets.Output()

# Function to add a scratch to the image
def add_scratch(image, x1, y1, x2, y2):
    draw = ImageDraw.Draw(image)
    draw.line((x1, y1, x2, y2), fill=(0, 0, 0, 255), width=3)

# Function to add a scuff to the image
def add_scuff(image, x, y):
    draw = ImageDraw.Draw(image)
    draw.point((x, y), fill=(0, 0, 0, 255))

# Function to simulate paint chipping
def add_paint_chipping(image, x, y, w, h):
    draw = ImageDraw.Draw(image)
    draw.rectangle((x, y, x+w, y+h), fill=(255, 255, 255, 255))

# Function to display image with defects
def show_image_with_defects(x1=0, y1=0, x2=0, y2=0, scuff_x=0, scuff_y=0, chip_x=0, chip_y=0, chip_w=10, chip_h=10):
    output.clear_output(wait=True)
    
    image_with_defects = original_image.copy()
    
    add_scratch(image_with_defects, x1, y1, x2, y2)
    add_scuff(image_with_defects, scuff_x, scuff_y)
    add_paint_chipping(image_with_defects, chip_x, chip_y, chip_w, chip_h)
    
    with output:
        plt.imshow(np.array(image_with_defects))
        plt.axis('off')
        plt.show()

# Create interactive widgets
interact(show_image_with_defects,
         x1=widgets.IntSlider(min=0, max=original_image.width, step=1, value=0),
         y1=widgets.IntSlider(min=0, max=original_image.height, step=1, value=0),
         x2=widgets.IntSlider(min=0, max=original_image.width, step=1, value=0),
         y2=widgets.IntSlider(min=0, max=original_image.height, step=1, value=0),
         scuff_x=widgets.IntSlider(min=0, max=original_image.width, step=1, value=0),
         scuff_y=widgets.IntSlider(min=0, max=original_image.height, step=1, value=0),
         chip_x=widgets.IntSlider(min=0, max=original_image.width, step=1, value=0),
         chip_y=widgets.IntSlider(min=0, max=original_image.height, step=1, value=0),
         chip_w=widgets.IntSlider(min=1, max=50, step=1, value=10),
         chip_h=widgets.IntSlider(min=1, max=50, step=1, value=10))

display(output)
```

# Proto4

```python
from PIL import Image, ImageDraw, ImageFilter, JpegImagePlugin
from ipywidgets import interact, widgets
import numpy as np
import random
import matplotlib.pyplot as plt

# Load the original image
image_path = "./P180-IMG_4314.png"  # Replace this with your image path
original_image = Image.open(image_path).convert("RGBA")

def add_scratch(image, x1, y1, x2, y2):
    draw = ImageDraw.Draw(image)
    draw.line((x1, y1, x2, y2), fill=(0, 0, 0, 255), width=3)
    
def add_scuff(image, x, y):
    draw = ImageDraw.Draw(image)
    draw.point((x, y), fill=(0, 0, 0, int(255 * 0.8)))  # 20% lighter

def add_paint_chipping(image, x1, y1, x2, y2):
    draw = ImageDraw.Draw(image)
    draw.rectangle((x1, y1, x2, y2), fill=(255, 255, 255, 255))

def add_dent(image, x, y, r):
    draw = ImageDraw.Draw(image)
    draw.ellipse((x-r, y-r, x+r, y+r), fill=(128, 128, 128, 128))

def add_warp(image, x1, y1, x2, y2):
    draw = ImageDraw.Draw(image)
    draw.polygon([(x1, y1), (x2, y1), (x2, y2), (x1, y2)], fill=(128, 128, 128, 128))

def show_image(image):
    plt.imshow(np.array(image))
    plt.axis('off')
    plt.show()

def interactive_defects(scratch=False, scuff=False, paint_chipping=False, dent=False, warp=False):
    image = original_image.copy()
    
    if scratch:
        add_scratch(image, random.randint(0, image.width), random.randint(0, image.height), random.randint(0, image.width), random.randint(0, image.height))
    if scuff:
        add_scuff(image, random.randint(0, image.width), random.randint(0, image.height))
    if paint_chipping:
        add_paint_chipping(image, random.randint(0, image.width), random.randint(0, image.height), random.randint(0, image.width), random.randint(0, image.height))
    if dent:
        add_dent(image, random.randint(0, image.width), random.randint(0, image.height), random.randint(5, 15))
    if warp:
        add_warp(image, random.randint(0, image.width), random.randint(0, image.height), random.randint(0, image.width), random.randint(0, image.height))
        
    show_image(image)

interact(interactive_defects, scratch=widgets.Checkbox(value=False), scuff=widgets.Checkbox(value=False), paint_chipping=widgets.Checkbox(value=False), dent=widgets.Checkbox(value=False), warp=widgets.Checkbox(value=False));
```
